# Agregación de datos por categoría

```python
import numpy as np
import pandas as pd
```

```python
gender = ["Male", "Female"]
income = ["Poor", "Middle Class", "Rich"]
```

```python
n = 500

gender_data = []
income_data = []

for i in range(0,n): # Genera n valores aleatorios para gender e income
    gender_data.append(np.random.choice(gender)) # Selecciona aleatoriamente un valor dentro de la lista gender
    income_data.append(np.random.choice(income)) # Selecciona aleatoriamente un valor dentro de la lista income
```

```python
gender_data[1:10]
```

```python
income_data[1:10]
```

```python
#Z -> N(0,1)
#N(m, s) -> m + s * Z
height = 180 + 30 * np.random.randn(n)
weight = 65 + 25 * np.random.randn(n)
age = 30 + 12 * np.random.randn(n)
income = 18000 + 3500 * np.random.rand(n)
```

```python
data = pd.DataFrame(
    {
        "Gender" : gender_data,
        "Economic Status" : income_data,
        "Height" : height,
        "Weight" : weight,
        "Age" : age,
        "Income" : income
    }
)
```

```python
data.head()
```

## Agrupación de datos
* Siempre tiene que ser sobre una variable categorica

```python
grouped_gender = data.groupby("Gender") 
```

```python
grouped_gender.groups
```

```python
i=0
for names, groups in grouped_gender:
    i+=1
    print (i)
    print(names) # Nombre del grupo (Female o Male)
    print(groups) # Datos asociados a cada grupo (Valores del resto de columnas para cada uno de los grupos)
```

```python
grouped_gender.get_group("Female")
```

```python
double_group = data.groupby(["Gender", "Economic Status"])
```

```python
len(double_group)
```

```python
for names, groups in double_group:
    print(names)
    print(groups)
```

```python
double_group.aggregate(
    {
        "Income": np.sum, # Suma de los valores de la columna ingresos de esta doble agrupación
        "Age" : np.mean, # Media de los valores de la columna ingresos de esta doble agrupación
        "Height" : np.std # Desviación típica de los valores de la columna ingresos de esta doble agrupación
    }
)
```

```python
double_group.aggregate(
    {
        "Age" : np.mean,
        "Height" : lambda h:(np.mean(h))/np.std(h) # Tipificación 
    }
)
```

```python
double_group.aggregate([np.sum, np.mean, np.std])
```

```python
double_group.aggregate([lambda x: np.mean(x) / np.std(x)])
```

## Dividir utilizando la distribución normal

```python
a = np.random.randn(len(data))
```

```python
len(a)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(a)
```

```python
check = (a<0.75) # De la distribución normal tomamos el 75% de los valores es lo que quiere decir a<0.75
```

```python
check # Es un array de boolean en el que se indica si es True que el valor era menor de 0.75 y False si era mayor o igual 0.75
```

```python
# Cuidado si quieres representar un vector de variables booleanas, 
# ya que al hacer un histograma obtendrás un error. 
# Ahora debes añadir un casting con .astype(int) al final de la lista de python para poderlo representar.
plt.hist(check.astype(int)) 
```

```python
training = data[check] # Con esta instrucción tomamos el 75% de las filas. 
                        # Las filas que se toman son las que corresponden a True que corresponden a la posición donde hay un True
testing = data[~check] # ~ se escribe con alt+126. 
                       # Negamos el vector booleano por lo que lo que antes era False ahora es True, aquí se selecciona el 25% de filas rstantes que corresponden a la posición donde hay un True
```
